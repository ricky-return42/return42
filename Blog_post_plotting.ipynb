```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def draw_circle(center,radius):
    # (X-center[0])**2 + (y-center[1])**2 = r**2
    x = np.linspace(center[0]-radius, center[0]+radius,100)
    posy = np.sqrt(radius**2 - (x-center[0])**2)
    negy = -posy 
    posy += center[1]
    negy += center[1]
    return x,posy,negy
x1,posy1,negy1 = draw_circle((0,0),4)
x2,posy2,negy2 = draw_circle((6,0),4)
x3,posy3,negy3 = draw_circle((3,27**0.5),4)
```

```python
sns.set_style('white')
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax.plot(x1,posy1, color='b', alpha=.6)
a1 = ax.plot(x1,negy1, color='b',alpha=.6,label='Domain Expertise')
ax.plot(x2,posy2, color='r',alpha=.6)
a2 = ax.plot(x2,negy2, color='r',alpha=.6,label='Hacking Skills')
a3 = ax.plot(x3,posy3, color='g',alpha=.6,label='Maths and Statistics Knowledge' )
ax.plot(x3,negy3, color='g',alpha=.6)
ax.set(xlim=[-5,11],ylim=[-5,10],xticks=[],yticks=[])
plt.gca().add_artist(plt.legend(handles=a1, loc=3))
plt.gca().add_artist(plt.legend(handles=a2, loc=4))
plt.legend(handles=a3, loc='upper center')
plt.show()
```
